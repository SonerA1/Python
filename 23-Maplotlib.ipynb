```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
ageList = [20,25,30,35,40,45,50,55,60]
```

```python
weightList = [70,74,70,75,89,90,76,83,92]
```

```python
ageList_numpy = np.array(ageList)
weightList_numpy = np.array(weightList)
```

```python
ageList_numpy
```

```python
weightList_numpy
```

```python
plt.plot(ageList_numpy,weightList_numpy,"r")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age-Weight Graph")
```

```python
plt.subplot(1,2,1) #1row, 2 column,1.graph
plt.plot(my_numpy1,my_numpy2, "r--")
plt.subplot(1,2,2)
plt.plot(my_numpy2,my_numpy1, "g*")
```

```python
my_figure = plt.figure()
type(my_figure)
my_axes = my_figure.add_axes([0.1,0.1,0.5,0.5])
my_axes.plot(my_numpy1,my_numpy2,"g*")
my_axes.set_xlabel("X Data")
my_axes.set_ylabel("Y Data")
my_axes.set_title("Graph Title")
```

```python
my_figure2 = plt.figure()



my_axes3 = my_figure2.add_axes([0.1,0.1,0.9,0.9])
my_axes3.plot(my_numpy1,my_numpy2,"r*-")
my_axes3.set_xlabel("X Data Large")
my_axes3.set_ylabel("Y Data Large")
my_axes3.set_title("Large Graph Title")

my_axes2 = my_figure2.add_axes([0.2,0.4,0.3,0.3])
my_axes2.plot(my_numpy1,my_numpy2,"g*")
my_axes2.set_xlabel("X Data Small")
my_axes2.set_ylabel("Y Data Small")
my_axes2.set_title("Small Graph Title")
```

```python
(my_fig, my_axe) = plt.subplots()
my_numpy3 = np.linspace(0,15,20)
my_numpy4 = my_numpy3 ** 2
my_axe.plot(my_numpy3, my_numpy4, color="#CD621D", alpha=1)
my_axe.plot(my_numpy4, my_numpy3, color="#B81DCB", alpha=0.9)
```

```python
(new_fig, new_axe) = plt.subplots()
new_axe.plot(my_numpy3, my_numpy3 + 2, color ="#B133FF", linewidth = 2.14 )
new_axe.plot(my_numpy3, my_numpy3 + 3, color ="blue", linewidth = 1.14 )
new_axe.plot(my_numpy3, my_numpy3 + 4, color ="orange", linewidth = 3.14 ,linestyle="-.")
new_axe.plot(my_numpy3, my_numpy3 + 5, color ="red", linewidth = 3.14 ,linestyle="-.", marker ="o")
new_axe.plot(my_numpy3, my_numpy3 + 6, color ="black", linewidth = 3.14 ,linestyle="-.", marker ="+", markersize=8)
```
